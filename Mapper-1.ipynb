# Config

```python
#%matplotlib inline
data_path = '/Users/cc/work/decmeg/data/'

subject_id = 1
```

# Code

```python
import os
import mapper
import numpy as np
import matplotlib.pyplot as plt
from dmutil import TrainSubject
from IPython.core.display import Image

# s = TrainSubject(subject_id, timesel=times, colsel=sensors, data_path=data_path)
s = TrainSubject(subject_id, data_path=data_path)
s.normalize()
```

```python
selection = np.random.rand(297 * 375) < 0.05
X_f = s.X_f.reshape((-1, 306))[selection]
X_s = s.X_s.reshape((-1, 306))[selection]
mapperinputdata = np.concatenate((X_f, X_s)).astype('double')
# mapperinputdata = data[selection].astype('double')
times = np.tile(np.tile(np.linspace(-1, 1, 375), 297)[selection], 2)
f = times
```

```python
X_f.shape
```

```python
# Preprocessing:
# E.g. intrinsic metric
#k = ...
#eps = ...
#mapperinputdata = mapper.metric.intrinsic_metric(inputdata, k, eps)

metricpar = {'metric': 'euclidean'}
# f = mapper.filters.eccentricity(mapperinputdata,
#     metricpar=metricpar,
#     exponent=1.0)

cover = mapper.cover.cube_cover_primitive(intervals=30, overlap=50.0)
cluster = mapper.complete_linkage()
mapper_output = mapper.mapper(mapperinputdata, f,
    cover=cover,
    cluster=cluster,
    cutoff=None,
    metricpar=metricpar)
# cutoff = mapper.cutoff.first_gap(gap=0.1)
# cutoff = lambda x, y: 2
cutoff=mapper.cutoff.biggest_gap()
mapper_output.cutoff(cutoff, f, cover=cover)

mapper_output.draw_scale_graph()
plt.savefig('scale_graph.pdf')

minsizes = []
mapper_output.draw_2D(minsizes=minsizes)
plt.savefig('mapper_output.pdf')

plt.show()
os.system("printf '\a'") 
```
