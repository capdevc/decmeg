```python
import numpy as np
import matplotlib.pyplot as plt
from dmutil import TrainSubject
from collections import Counter
%matplotlib inline

def mergethrees(a):
    newcols = a.shape[2] / 3
    b = np.zeros((a.shape[0], a.shape[1], newcols))
    for i in range(newcols):
        b[:, :, i] = a[:, :, i * 3] + a[:, :, i * 3 + 1] + a[:, :, i * 3 + 2]
    return b


c = Counter()
f, axarr = plt.subplots(4, 4)
data_path = '/Users/cc/work/decmeg/data/'

for i in range(16):
    axx = i % 4
    axy = i / 4

    s = TrainSubject(i + 1, timesel=slice(125,None), data_path=data_path)
    s.normalize()

    # x_f = mergethrees(s.X_f)
    # x_s = mergethrees(s.X_s)
    x_f = s.X_f
    x_s = s.X_s

    avg_x_s = np.mean(x_s, axis=0)

    diffs = np.abs(x_f - avg_x_s)

    sums = np.sum(diffs, axis=0)

    hots = np.argsort(np.sum(sums, axis=0))[:-50:-1]
    # print np.sum(diffs, axis=0)
    print('Sensors for subject {0}: {1}'
          .format(i + 1, hots))
    c.update(hots)

    axarr[axx, axy].imshow(sums,
                           vmin=np.min(sums),
                           vmax=np.max(sums),
                           origin='lower',
                           cmap='gist_heat',
                           interpolation='nearest',
                           aspect='auto')
    axarr[axx, axy].set_title('Subject {0:0>2}'.format(i + 1))
    axarr[axx, axy].grid(b=None)

print(c.most_common(30))

ticks = []
for val, _ in c.most_common(30):
    ticks.append(val)

for i in range(16):
    axx = i % 4
    axy = i / 4
    colors = ['b'] * len(ticks)
    # axarr[axx, axy].vlines(ticks, 0, 40, colors='b', linestyles='solid')
    axarr[axx, axy].set_xticks(ticks)
    axarr[axx, axy].tick_params(axis='both',
                                direction='out',
                                labelsize='xx-small',
                                labelbottom=True)

plt.show()
```
